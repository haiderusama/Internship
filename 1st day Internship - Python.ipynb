```python
greeting = "Usama Haider"
greeting_space = "   Usama Haider    "
print(greeting.startswith("U"))
print(greeting.startswith("Usama"))
print(greeting_space.startswith(" "))
print(greeting.startswith("   "))
```

Slice

```python
Data = "From Usama Haider to Wasi Ur Rehman Qamar on wasiurrehmanqamar@ntu.edu.pk at junary 28,2023"
Hat = Data.find("@")
print(Hat)
Finding = Data.find(" ", Hat)
print(Finding)
email = Data[Hat+1 : Finding]
print(email)
```
